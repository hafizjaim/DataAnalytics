```python
# let's create a new column of the ratio of how many ball hit the target for every short and following other related column as well

df['OnTarget_shot'] = (df['on target'] / df['Shots '].sum()) * 100

df['Accurate_Passes'] = (df['accurate'] / df['Passes '].sum()) * 100

df['Accurate_Crosses'] = (df['accurate.1'] / df['Crosses '].sum()) * 100

df['Success_Dribble'] = (df['successful'] / df['Dribbles '].sum()) * 100

df['Won_Duels'] = (df['won'] / df['Duels '].sum()) * 100

df['Won_AerialDuels'] = (df['won.1'] / df['Aerial duels '].sum()) * 100

df['Accurate_ThroughPasses'] = (df['accurate.2'] / df['Through passes'].sum()) * 100

df['Accurate_PaassesToFinalThird'] = (df['accurate.3'] / df['Passes to final third'].sum()) * 100

df['Accurate_PaassesToPenaltyArea'] = (df['accurate.4'] / df['Passes to penalty area '].sum()) * 100

```
